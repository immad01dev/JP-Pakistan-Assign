# Add student grades column according to percentages



```python
#Import Libraries
import csv
```

```python
# Opening CSV File
open_csv = open('student data.csv',mode='r')
open_csv
```

```python
# Reading CSV File
read_csv = csv.reader(open_csv)
read_csv
```

```python
# Displaying CSV Data
data = list(read_csv)
print(data)
```

```python
# Adding Column Total
data[0].append('Total')
data[0].append('Grade')
```

```python
data[0]
```

```python
# Iterating Displaying Data and Calculating Total Marks and Grade
for i in data[1:]:
  chem = float(i[2]);
  phy = float(i[3]);
  math = float(i[4]);

  total = round((chem + phy + math)/300*100,2)

  i.append(total)

  if total >=80 and total <= 100:
    i.append('A')

  elif total >= 70 and total <= 79:
    i.append('C')

  elif total >= 60 and total <= 69:
    i.append('D')

  elif total >= 50 and total <= 59:
    i.append('F')

  else:
    i.append('Fail')

print(data)
```

```python
# Writing Updated Data into CSV
updated = open('student data.csv',mode='w',newline='')

csv_write = csv.writer(updated)
csv_write.writerows(data)

updated.close()
```
