### **Importing Libraries & Reading CSV**

```python
import pandas as pd


df = pd.read_csv('Salaries.csv')
df.head(5)
```

## Assignment Questions

Drop rows where EmployeeName and JobTitle are 'Not provided'

```python
find_rows = df.EmployeeName.str.contains('Not provided') & df.JobTitle.str.contains('Not provided')
df[~find_rows]
```

Replace 'Not Provided' with 0 in BasePay OvertimePay	OtherPay	Benefits

```python
# df.Basepay = df.BasePay.replace('Not provided',0)
# df.OvertimePay = df.OvertimePay.replace('Not provided',0)
# df.OtherPay = df.BasePay.replace('Not provided',0)

df.replace('Not Provided',0).BasePay
df.replace('Not Provided',0).OvertimePay
df.replace('Not Provided',0).OtherPay

print(df.BasePay)
print(df.OvertimePay)
print(df.OtherPay)
```

Change BasePay	OvertimePay	OtherPay	Benefits type to float

```python
df.BasePay = df.BasePay.replace('Not Provided',0).astype(float)
df.OvertimePay = df.OvertimePay.replace('Not Provided',0).astype(float)
df.Benefits = df.Benefits.replace('Not Provided',0).astype(float)

print(df.BasePay)
print(df.OvertimePay)
print(df.OtherPay)
```
